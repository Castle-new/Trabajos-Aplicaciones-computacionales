```python
import numpy as np
import matplotlib.pyplot as plt

# Genero la matriz 100x100 con números aleatorios de distribución normal
matriz = np.random.normal(loc=0, scale=1, size=(100, 100))

# Con la función linalg.svd realizo la descomposición en valores singulares
U, S, Vt = np.linalg.svd(matriz)
#print(matriz)
#print(S)

# Grafico los valores singulares (ya están en orden descendente)
plt.plot(np.arange(0, len(S)), S, marker='o', linestyle='-')
plt.xlabel('Número de valor singular (n)')
plt.ylabel('Valor singular')
plt.title('Valores singulares en orden descendente')
plt.grid(True)
plt.show()
```

# Se observa que el comportamiento es similar al de una recta con pendiente negativa.

```python
num_experimentos = 100
num_valores_singulares = 10

# Almacena los primeros 10 valores singulares de cada experimento
valores_singulares_experimentos = []

# experimentos
for _ in range(num_experimentos):
    # matriz 100x100 aleatoria de distribución normal
    A = np.random.normal(loc=0, scale=1, size=(100, 100))
    
    # Calcula la SVD
    _, s, _ = np.linalg.svd(A)
    
    # Guarda primeros 10 valores singulares
    valores_singulares_experimentos.append(s[:num_valores_singulares])

# Convierte la lista de vectores en un array para facilitar el manejo
valores_singulares_experimentos = np.array(valores_singulares_experimentos)

#print(valores_singulares_experimentos)

plt.figure(figsize=(8, 5))
plt.boxplot(valores_singulares_experimentos, patch_artist=True)
plt.xlabel('Número de valor singular (n)')
plt.ylabel('Valor singular')
plt.title('Distribución de los primeros 10 valores singulares')
plt.grid(True)
plt.show()

# Estadísticas
for i in range(num_valores_singulares):
    media = np.mean(valores_singulares_experimentos[:, i])
    minimo = np.min(valores_singulares_experimentos[:, i])
    maximo = np.max(valores_singulares_experimentos[:, i])
    
    print(f"\nEstadísticas para el valor singular {i+1}:")
    print(f"  Media: {media}")
    print(f"  Valor mínimo: {minimo}")
    print(f"  Valor máximo: {maximo}")
```
